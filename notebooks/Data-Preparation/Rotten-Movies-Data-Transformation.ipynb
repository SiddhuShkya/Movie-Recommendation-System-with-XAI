```python
# Importing Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_movies = pd.read_csv("../../data/movie-review-data/rotten_tomatoes_movies.csv")
df_movies.head(2).T
```

```python
null_percentage = df_movies.isnull().mean() * 100
null_percentage_df = null_percentage.sort_values(ascending=False).to_frame(name='MissingPercent')
null_percentage_df
```

```python
df_movies.drop(columns=['audienceScore', 'tomatoMeter', 'rating', 'ratingContents', 'releaseDateTheaters', 'boxOffice', 'distributor', 'soundMix'], inplace=True)
df_movies.sample(3).T
```

```python
df_movies['genre'].value_counts()
```

```python
df_movies['genre'] = df_movies['genre'].apply(
    lambda x: [g.strip().lower() for g in x.split(",")] if isinstance(x, str) else x
)
df_movies['genre'].value_counts()
```

```python
df_movies.to_csv('../../data/movie-review-data/transformed_movies.csv', index=False)
```
